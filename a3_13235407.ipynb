**QUESTION**

Following your graduation, you are hired by a polling organisation as a data analyst.
As social media has exploded and transformed the way people interact with each
other, it would be a great idea to use messages collected from social media to
predict how the user can be converted to change his/her support. List three
challenges to solving this problem. With reference to existing approaches, describe
the design of your system. Discuss the ethical and social consequence of this
study.

**INTRODUCTION**

Nowadays, in Australia and in most of the countries in the world, people live in an interconnected society where it is possible and easy for everyone to share their opinions with a large number of other users without any restriction and filter thanks to the use of social media.  It is possible to understand the kind of impact that Social Media are having on the society by reading the numbers of active users that the most popular platforms have each month. Facebook and YouTube both have more than 2 billion single users per month, while QQ and Instagram have respectively 820 million and 1 billion visitors.

While many studies are trying to understand the social effects that the possibility of free speech for so many people is having on the society, the pre-mentioned companies find it hard to control hate speech and the topics that are discussed on their platforms. 
This situation allows the possibility of capturing what people really feel and think about most of the topics, giving also the chance to understand what are the themes that mostly affect the society at the moment.

This report explores the possibility of analysing the contents that users post on social media to increase the consensus of a potential political candidate during the elections.


**BACKGROUND**

Given the fact that nowadays people use social media to express their opinions on almost every aspect of their life, it is common to read about people’s personal political believes, especially on platforms like Facebook and Twitter. In addition, multiple studies have showed that people expressing their political opinions on Facebook are more likely to be engaged in real life politic. Because of this, it is crucial to analyse what people post on social media and understand their point of view to intercept their votes.

**CHALLENGES**

A project that analyses people’s posts to increase the support of a single candidate can bring with itself multiple issues of technical and ethical nature. 
This report will explore and try to solve three different challenges of different nature.

**API and code**

The first challenge to face when intending to analyse messages coming from social media is the technique used to obtain those messages.
Most companies nowadays offer APIs to the developers that intend to work with their platforms; using a social media APIs is probably the best method to collect messages from the platforms without incurring in illegal activities. In this way the privacy of the users would still be protected and the developer would not break the rules of the social network.
It is also important to develop a code that analyses the messages collected and outputs clear information on people’s ideas.

**What to analyse**

It is important to identify the right contents to collect and make sure to use them in the right way. 
Taking, once again, Facebook and Twitter as examples this paragraph will try to understand what are the best contents to analyse.
Facebook could be used to analyse comments and interactions on political opponents’ pages and profiles in order to understand whether a particular topic attracts the interest of people (voters) or not. By analysing the comments and how many people shared single posts it is possible to have an idea of what people think of the subject. Each element should then be compared to other posts from the same account to better understand if the topic can be considered viral. 
Users on Twitter, on the other side, do a large use of hashtags and this makes it easy to understand what contents are viral in real time; this could be used to increase political consent by expressing a popular idea each time a topic is going viral.


**Extremist messages capture more attention**

A challenge that a project like the one that is being discussed in this report would probably face regards hate speech and extremist messages. Currently, it is not uncommon to read messages containing hate speech on social media. This kind of messages attract the media and other users’ attention much more than messages that contain pacific or politically correct opinion.
The fact that this kind of messages and ideas are more popular than their counterparts on social media does not mean that in a real life scenario, such as the elections, they would conquer more votes. Therefore, it is important to calibrate the ideas by using data from previous elections in order to make a correct prevision.



**DESIGN**

The desired product would collect messages from social media like Facebook and Twitter as inputs and output topics that a potential candidate should discuss in order to obtain people’s attention and therefore, their votes. 
This technique is already used by most politicians nowadays, in fact, it is common, especially during election campaigns, to see politicians mention and take positions about viral contents even if these are not related anyhow to politics.


**CONSEQUENCES**

A project like the one discussed in this report would bring unavoidable ethical consequences. The fact that a politician would not be driven by his own ideas but from the ideas that would potentially give him a bigger number of votes could be considered as unreliable and as someone who does not do politics to improve the society but only for his/her own power. 

**CONCLUSION**

To conclude, in order to increase his/her support, a politician could use data and messages collected from social media through the use of APIs and other technologies, to predict what topics and ideas that would make people vote for him/her. This idea, already used by multiple politicians, would bring unavoidable ethical consequences such as a questionable personal integrity of the user.
